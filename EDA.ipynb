```python
df.select_dtypes(include="number").corr()["Price (in rupees)"].sort_values(ascending=False)
```

```python
df["Carpet Area"] = (
    df["Carpet Area"]
    .str.replace(" sqft", "", regex=False)
)

df["Carpet Area"] = pd.to_numeric(df["Carpet Area"], errors="coerce")
```

```python
df["Super Area"] = (
    df["Super Area"]
    .str.replace(" sqft", "", regex=False)
)

df["Super Area"] = pd.to_numeric(df["Super Area"], errors="coerce")
```

```python
df["Floor"] = df["Floor"].str.extract(r"(\d+)").astype(float)
```

```python
df["Bathroom"] = pd.to_numeric(df["Bathroom"], errors="coerce")
df["Balcony"] = pd.to_numeric(df["Balcony"], errors="coerce")
```

```python
df.select_dtypes(include="number")\
  .corr()["Price (in rupees)"]\
  .sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Price (in rupees)'],bins=30)
plt.xlabel("price")
plt.ylabel("count")
plt.show()
```

```python
import seaborn as sns
sns.boxplot(x=df["Price (in rupees)"])
plt.title("Boxplot of Price")
plt.show()

```

```python
import numpy as np

sns.histplot(np.log(df["Price (in rupees)"]), bins=30, kde=True)
plt.title("Log-Transformed Price Distribution")
plt.show()
```
