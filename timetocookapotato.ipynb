```python
import pandas as pd

# Create dataset
data = {
    'Power_Level': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'Cook_Time': [20, 13, 9.5, 7.2, 6.2, 6.0, 6.1, 6.5, 7.2, 8.2]
}

df = pd.DataFrame(data)


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df[['Power_Level']]
y=df['Cook_Time']
```

```python
poly=poly=PolynomialFeatures(degree=4)
X_poly=poly.fit_transform(X)
```

```python
print(X_poly)
```

```python
model=LinearRegression()
model.fit(X_poly,y)
```

```python
y_pred = model.predict(X_poly)

```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MSE:", mse)
print("R2 Score:", r2)
```

```python
plt.scatter(X, y, color='red', label='Actual points')
plt.plot(X, y_pred, color='blue', label='Polynomial Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression Example')
plt.legend()
plt.show()
```
